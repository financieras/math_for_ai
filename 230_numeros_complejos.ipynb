# Números complejos
- Están formados por una parte real y una parte imaginaria.

Los números reales son los que ya conocemos. Abarcan los siguietes:
1.  Números naturales: 1, 2, 3, 4, ...
2. Números enteros: Los naturales más los negativos, más el cero: ... -3, -2, -1, 0, 1, 2, 3, ...
3. Números racionales: Se pueden poner como a/b: -5/3, 0, 4/2, 19/7
4. Números irracionales: Con infinitos decimales: π, e, $\sqrt{2}$, $\phi$, ...

Los **números imaginarios** surgen al intentar resolver una ecuación de este tipo:
$$x^2=-1 \quad \Rightarrow \quad x = \sqrt{-1}$$

A la raiz cuadrada de -1 se la llama i que es la unidad imaginaria:

$$i = \sqrt{-1}$$

### Ejemplos de números complejos
- $7 + 2i$
- $-3 + 5i$
- $4 - 4i$
- $-1 - 7i$
- $2.2 + 3.3i$
- $0 + 5i = 5i$
- $7 + 0i = 7$ (cualquier número real se puede interpretar como un complejo con parte imaginaria cero)
- $\frac{1}{3} + \sqrt{2}i$

## En Python
En ingeniería eléctrica la letra $i$ se reservaba para designar la corriente eléctrica (intensidad) por lo que la unidad imaginaria se designaba como $j$ para no confundir. Esta es la notación que se ha seguido en Python.

- En Python la unidad imaginaria es $j =  \sqrt{-1}$.
- No se necesita ninguna librería para trabajar con complejos, usando el tipo integrado `complex`.
