# Funciones

## 1. Introducción a las funciones

### 1.1. Concepto de función como relación entre variables
En matemáticas una **función** es la relación que hay entre una magnitud y otra.

Si una magnitud la representamos por la variable $x$ y la otra magnitud la representamos por la variable $y$, diremos que la función $f$ es la que relaciona la variable independiente $x$ con la variable dependiente $y$. Se representa como $y=f(x)$ y se lee diciendo que $y$ es función de $x$.

**Ejemplo**
Podemos observar como la temperatura varía a lo largo del día. Podríamos establecer una función que relacione la temperatura con la hora del día.
- La variable independiente sería la hora del día.
- La variable dependiente sería la temperatura.

**Variables**
- Variable independiente $x$: es el valor que podemos elegir o que varía libremente.
- Variable dependiente $y$: es el valor que se obtiene como resultado.


¿Qué es una función?

- Es como una máquina que "transforma" o "procesa" valores de entrada para producir valores de salida.

-  Una tabla de valores `xy` es una representación de datos, mediante pares ordenados, que expresan la relación existente entre dos magnitudes o dos situaciones.

<img src="https://github.com/financieras/math_for_ai/blob/main/img/funcion_maquina2.png?raw=1" alt="función como máquina" width="480"/>

- La variable $x$ es la variable independiente. La variable $y$ es la variable dependiente.

**Otros ejemplos**

- La edad de una persona y su altura
- El lado de un cuadrado y su área
- Los kilómetros recorridos y el consumo de energía
- El precio a pagar según la cantidad de kilogramos comprados de fruta


**Representación gráfica de funciones**

- El plano cartesiano como herramienta de visualización
    - Eje horizontal $x$ para la variable independiente
    - Eje vertical $y$ para la variable dependiente
    - Cada punto del plano $(x,y)$ representa una pareja de valores relacionados
- Visualización de la correspondencia única
    - Trazar líneas verticales en cualquier punto del eje x
    - Una función debe cortar esta línea vertical en un único punto
    - Si corta en más de un punto, no es una función

**Representacón gráfica**
- Regla de la línea vertical: Nos permite identificar si una gráfica es o no función. Toda vertical trazada desde cualquier punto $x$ debe cortar la recta en como máximo un solo punto. Si la recta vertical corta en dos o más puntos a la gráfica no es una función.

**Dominio y recorrido**
- El subconjunto de $\mathbb{R}$ para los cuales está definida la función, se denomina dominio de $f$, y se denota $Df$
- Los valores que toma la imagen conforman un subconjunto denominado imagen, rango o recorrido de $f$, y se denota $Rf$.
- En términos más simples, el dominio corresponde a lo que yo manejo, es decir de donde yo vengo, y el recorrido es hacia donde yo voy.

## Gráfica de una función con `matplotlib`
- Primero necesitamos generar el rango de variación de la $x$ para el eje horizontal.
- Podemos hacerlo de dos formas (preferimos la segunda):
    1. Usando np.arange(inicio, fin, paso). El fin queda excluido.
    2. Usando np.linspace(inicio, fin, número de puntos)

```python
import numpy as np
x = np.arange(0,20,2)   # no llega hasta 20, para llegar poner 22
x
```

```python
import numpy as np
x = np.linspace(0,25,101)
x
```

```python
#@title Gráficar el Polinomio $ y = x^3 - 3x^2 + x + 1 $

import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-5, 5, 101)

f = x**3 - 3 * x**2 + x + 1

plt.figure()
plt.plot(x, f)
plt.grid()

plt.show()
```

```python
#@title Graficar la función $y=3e^{-2x^2} \sin(x)$

import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-5, 5, 101)

f = 3 * np.exp(-2*x**2) *  np.sin(x)

plt.figure()
plt.plot(x, f)
plt.grid()

plt.show()
```

```python
#@title Graficar la función $y=\ln(x) -x + 2$

import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(0.01, 5, 101)

f = 3 * np.log(x) -x + 2
plt.figure()
plt.plot(x, f)
plt.grid()

plt.show()
```

```python
#@title Graficando las tres funciones simultáneamente

import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-1, 4, 101) # con log sale un warning ya que no está definido para valores negativos o cero

f = x**3 - 3 * x**2 + x + 1
g = 3 * np.exp(-2*x**2) *  np.sin(x)
h = 3 * np.log(x) -x + 2
p = 2 * x - 5

plt.figure()
plt.plot(x, f)
plt.plot(x, g)
plt.plot(x, h)
plt.grid()

plt.show()
```

```python
#@title Graficando dos funciones con etiquetas

import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-1, 4, 101)

f = x**3 - 3 * x**2 + x + 1
g = 3 * np.exp(-2*x**2) *  np.sin(x)


plt.figure()
plt.plot(x, f)
plt.plot(x, g)
plt.legend(['polinomio', 'exponencial'])
plt.title('Funciones')
plt.xlabel('eje x')
plt.ylabel('eje y')

plt.grid()

plt.show()
```

```python
#@title Graficando dos funciones una a cada lado

import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-2.1, 4, 100)

f = x**3 - 3 * x**2 + x + 1
g = 3 * np.exp(-2*x**2) *  np.sin(x)


plt.figure()

plt.subplot(1,2,1)  # 1 fila y 2 columnas de funciones, esta es la función num. 1
plt.plot(x, f)
plt.grid()

plt.subplot(1,2,2)  # 1 fila y 2 columnas de funciones, esta es la función num. 2
plt.plot(x, g)
plt.grid()

plt.show()
```

```python
#@title Graficando las cuatro funciones

import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(0.01, 4, 100)

f = x**3 - 3 * x**2 + x + 1
g = 3 * np.exp(-2*x**2) *  np.sin(x)
h = 3 * np.log(x) -x + 2
p = 2 * x - 4

plt.figure()

plt.subplot(2,2,1)  # 2 filas y 2 columnas de funciones, esta es la función num. 1
plt.plot(x, f)
plt.grid()

plt.subplot(2,2,2)  # 2 filas y 2 columnas de funciones, esta es la función num. 2
plt.plot(x, g)
plt.grid()

plt.subplot(2,2,3)  # 2 filas y 2 columnas de funciones, esta es la función num. 3
plt.plot(x, h)
plt.grid()

plt.subplot(2,2,4)  # 2 filas y 2 columnas de funciones, esta es la función num. 4
plt.plot(x, p)
plt.grid()

plt.show()
```

```python
#@title Pocos puntos para graficar la función seno

import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(0, 10, 8)   # con solo 8 puntos no se ve bien la función seno

y = np.sin(x)

plt.plot(x, y)

plt.show()
```

```python
#@title Graficando la función seno con etiquetas

import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(0, 10, 101)

y = np.sin(x)

plt.plot(x, y, label='y=sin(x)')

plt.xlabel('x en radianes')
plt.ylabel('y')
plt.title('Gráfico de la función seno')
plt.legend()    # toma el valor de la label indicada en plot
plt.xticks(np.arange(0, 10+1, np.pi/4)) # indica el grid vertical entre 0 y 10 cada pi/4
plt.yticks(np.arange(-1, 1.1, .5))      # indica el grid horizontal entre -1 y 1 cada 0.5

plt.grid(True)

plt.show()
```

```python
#@title Graficando la función de activación sigmoide

import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-10, 10, 101)

y = 1/(1+np.exp(-x))

plt.plot(x, y, label='y=1/(1+e^(-x))')

plt.xlabel('x')
plt.ylabel('y (probabilidad entre 0 y 1)')
plt.title('Gráfico de la función sigmoide')
plt.legend()    # toma el valor de la label indicada en plot

plt.grid()

plt.savefig("sigmoide.png", dpi=150) # para grabar el gráfico en un archivo

plt.show()
```

```python
#@title Función escalón

import numpy as np
import matplotlib.pyplot as plt

def escalon(t):
    return np.piecewise(t, [t < 0, t >= 0], [0, 1])

t = np.linspace(-5, 5, 1000)
y = escalon(t)

plt.figure(figsize=(5, 3))
plt.plot(t, y, 'b-', linewidth=2)
plt.title('Función Escalón Unitario')
plt.xlabel('t')
plt.ylabel('u(t)')
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
#@title Función de activación ReLu (Rectified Linear Unit)

import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

x = np.linspace(-10, 10, 1001)
y = relu(x)

plt.figure(figsize=(5, 3))
plt.plot(x, y, 'b-', linewidth=2, label='y = max(0, x)')
plt.title('Función de Activación ReLU')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.ylim(-0.1, 10)
plt.legend()
plt.show()
```

```python
#@title Función de activación Tanh (tangente hiperbólica)

import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

x = np.linspace(-5, 5, 1001)
y = tanh(x)

plt.figure(figsize=(5, 3))
plt.plot(x, y, 'b-', linewidth=2)
plt.title('Función de Activación Tanh')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.ylim(-1.1, 1.1)
plt.show()
```
